## Funciones


```python
def separarInfo(df):
  df['Fecha de Pedido'] = pd.to_datetime(df['Fecha de Pedido'])
  df['Hora'] = df['Fecha de Pedido'].dt.hour
  df['Dia'] = df['Fecha de Pedido'].dt.day
  df['Mes'] = df['Fecha de Pedido'].dt.month

  # Separamos el dia de la semana siendo lunes 0 y domingo 6
  df['Dia de la Semana'] = df['Fecha de Pedido'].apply(lambda x : x.weekday())

  # Separamos la direccion de envio en calle ciudad y PO BOX
  dfDireccion = df['Dirección de Envio'].str.split(', ',expand=True)
  dfDireccion.columns = ['Calle','Ciudad','PO BOX']

  # Separamos el PO BOX en Estado y Cod Postal
  dfDireccion[['Estado', 'Cod Postal']] = dfDireccion['PO BOX'].str.split(' ',expand=True)
  df = pd.concat([df,dfDireccion],axis=True)

  return df
```

## Preparación de los dataset

```python
rutaBase = '/content/drive/MyDrive/Materias/Big Data/Practico1/Dataset de ventas2'

archivos = os.listdir(rutaBase)
dataframes = {}

for archivo in archivos:
  if archivo.endswith('.csv'):
    # Preparacion key/value
    mes = archivo.split('_')[3].split('.')[0]
    rutaCompleta = os.path.join(rutaBase, archivo)
    df = pd.read_csv(rutaCompleta)
    dfVentas = df.copy()

    # Limpieza y creacion de columnas
    dfVentas['Cantidad Pedida'] = pd.to_numeric(dfVentas['Cantidad Pedida'], errors='coerce')
    dfVentas['Precio Unitario'] = pd.to_numeric(dfVentas['Precio Unitario'], errors='coerce')

    dfVentas = dfVentas.dropna()

    dfVentas['Cantidad Pedida'] = dfVentas['Cantidad Pedida'].astype(int)
    dfVentas['Precio Unitario'] = dfVentas['Precio Unitario'].astype(float)

    dfVentas = separarInfo(dfVentas)

    # Carga del df al diccionario
    dataframes[mes] = dfVentas

dataframes['Enero']
```

```python
MESES = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
SEMANA = ['Lunes', 'Martes', 'Miercoles', 'Jueves', 'Viernes', 'Sabado', 'Domingo']
HORAS = list(range(24))
```

### Ventas totales por mes y total recaudado por mes

```python
ventasTotalesMes = []
recaudadoTotalMes = []

for mes in MESES:
  ventasTotalesMes.append(dataframes[mes]['Cantidad Pedida'].sum())
  recaudadoTotalMes.append(int((dataframes[mes]['Cantidad Pedida'] * dataframes[mes]['Precio Unitario']).sum()))

fig = go.Figure()

fig.add_trace(go.Bar(
  x=MESES,
  y=ventasTotalesMes,
  name='Ventas totales por mes',
  marker_color='indianred',
  text=ventasTotalesMes,
  textposition='auto',
  yaxis='y',
  offsetgroup=1
))

fig.add_trace(go.Bar(
  x=MESES,
  y=recaudadoTotalMes,
  name='Recaudado total por mes',
  marker_color='lightsalmon',
  text=recaudadoTotalMes,
  textposition='auto',
  yaxis='y2',
  offsetgroup=2
))

fig.update_layout(
  barmode='group',
  bargap=0.1,
  xaxis=dict(title='Mes'),
  legend=dict(
    x=0,
    y=1.0,
    bgcolor='rgba(255, 255, 255, 0)'
  ),
  yaxis=dict(title='Total de unidades vendidas'),
  yaxis2 = dict(title='Total recaudado por mes', overlaying= 'y',side='right'),
  yaxis_range = [0, max(ventasTotalesMes)*1.1],
  yaxis2_range = [0, max(recaudadoTotalMes)]
)

fig.show()
```

### Ventas totales de cada producto por mes

```python
# Creamos un diccionario que almacenara acada nombre de los productos y su valor sera otro diccionario con la cantidad vendida de ese producto en cada mes
productos = {}

for mes in MESES:

  ventasMes = dataframes[mes].groupby('Producto')['Cantidad Pedida'].sum()
  productosMes = ventasMes.reset_index(inplace=False)['Producto'].values

  for producto in productosMes:
    if not(producto in productos):
      productos[producto] = {'Enero':0, 'Febrero':0, 'Marzo':0, 'Abril':0, 'Mayo':0, 'Junio':0, 'Julio':0, 'Agosto':0, 'Septiembre':0, 'Octubre':0, 'Noviembre':0, 'Diciembre':0}
      productos[producto][mes] = ventasMes.loc[producto]
    else:
      productos[producto][mes] = ventasMes.loc[producto]

fig = go.Figure()

# Creamos un grafico por cada producto
for producto in productos:
  fig.add_trace(go.Scatter(
    x=MESES,
    y=list(productos[producto].values()),
    mode='lines',
    name=producto,
    line=dict(width=2, shape='spline'),
  ))

fig.update_layout(
  title='Unidades vendidas de cada producto por mes',
  xaxis=dict(title='Meses', tickvals=MESES),
  yaxis=dict(title='Unidades vendidas', side='left'),
  legend=dict(y=0.3, yanchor='bottom'),
)

fig.show()
```

## 2. Optimización de la publicidad y patrón de ventas por hora:
* ¿Cuál es el momento ideal del día para presentar la publicidad y aumentar la
probabilidad de compra?
* ¿Cómo cambian los patrones de ventas por hora a lo largo del año?
* ¿Hay modificaciones en los patrones de ventas durante las horas de mayor actividad en los distintos meses?


```python
# Creamos un diccionario para guardar los valores de cada hora y una lista para guardar los promedios
ventasPorHora = {}
promedios = [0] * 24

# Iteramos para obtener los totales vendidos por hora de cada mes
for mes in MESES:
  promedioHoras = dataframes[mes].groupby('Hora')['Cantidad Pedida'].sum()
  ventasPorHora[mes] = promedioHoras.values

# Iteramos para obtener los promedios de cada mes en cada hora
for venta in ventasPorHora:
  for i in range(len(promedios)):
    promedios[i] += ventasPorHora[venta][i]

for i in range(len(promedios)):
  promedios[i] = promedios[i] / len(ventasPorHora)

fig = go.Figure()

# Creamos un grafico por cada mes, para ver la diferencia entre ellos a lo largo de las horas
for venta in ventasPorHora:
  fig.add_trace(go.Scatter(
    x=HORAS,
    y=ventasPorHora[venta],
    mode='lines',
    name=venta,
    line=dict(width=2, shape='spline')
  ))

# Creamos un grafico extra para ver el promedio de todos los meses en cada hora
fig.add_trace(go.Scatter(
  x=HORAS,
  y=promedios,
  mode='markers',
  name='Promedio Anual',
  line=dict(width=4, shape='spline', color='black')
))

fig.update_layout(
  title='Unidades vendidas por hora en cada mes',
  xaxis=dict(title='Horas', tickvals=HORAS),
  yaxis=dict(title='Unidades vendidas', side='left'),
  legend=dict(y=0.3, yanchor='bottom')
)

fig.show()
```

### Ventas por ciudad y Estado


```python
# Creamos un dataframe para almacenar la informacion de los estados y ciudades
dfVentasEstado = pd.DataFrame()

for mes in MESES:
  ventasEstado = dataframes[mes].groupby(['Estado','Ciudad'])['Cantidad Pedida'].sum()
  ventasEstado = ventasEstado.reset_index(inplace=False)

  if dfVentasEstado.empty:
    dfVentasEstado = ventasEstado
  else:
    # Hacemos que se sume el valor actual de la cantidad pedida con la cantidad previamente cargada ene el dataset para cada codigo postal
    dfVentasEstado.loc[dfVentasEstado['Ciudad'] == ventasEstado['Ciudad'], 'Cantidad Pedida'] += ventasEstado.loc[dfVentasEstado['Ciudad'] == ventasEstado['Ciudad'], 'Cantidad Pedida']

# Creamos un dataframe que almacenara la informacion con detalles para mostrarla de mejor manera en el grafico
dfVentasEstadoDetallado = dfVentasEstado.groupby('Estado')['Cantidad Pedida'].sum()
dfVentasEstadoDetallado = dfVentasEstadoDetallado.reset_index(inplace=False)
dfVentasEstadoDetallado['Detalle Ciudad'] = ''

# Adaptamos la informacion para que no tenga conflictos con el grafico
dfVentasEstado['Cantidad Pedida str'] = dfVentasEstado['Cantidad Pedida'].apply(lambda x : str(x))
dfVentasEstado['Detalle Ciudad'] = dfVentasEstado['Ciudad'] + ' - ' + dfVentasEstado['Cantidad Pedida str']

for fila in dfVentasEstado.values:
  dfVentasEstadoDetallado.loc[dfVentasEstadoDetallado['Estado'] == fila[0], 'Detalle Ciudad'] += fila[4] + ' | '

fig = px.choropleth(
  dfVentasEstadoDetallado,
  locations='Estado',
  locationmode='USA-states',
  color='Cantidad Pedida',
  color_continuous_scale="RdYlGn",
  scope='usa',
  title= 'Ventas por Estado y Ciudades',
  hover_name=dfVentasEstadoDetallado['Estado'] + ' / ' + dfVentasEstadoDetallado['Detalle Ciudad']
)

# Añadimos anotaciones para dar la informacion de las ventas totales por ciudad
fig.add_annotation(
  text='Ventas totales por Ciudad:',
  xref='paper',
  yref='paper',
  x=0,
  y=1,
  showarrow=False
)

for i in range(len(dfVentasEstado['Detalle Ciudad'])):
  fig.add_annotation(
    text=dfVentasEstado['Detalle Ciudad'].iloc[i] + ' / ' + dfVentasEstado['Estado'].iloc[i],
    xref='paper',
    yref='paper',
    x=0,
    y=i/10,
    showarrow=False
  )

fig.show()
```

### Variación en las ventas por estado a lo largo de los meses

```python
# Creamos un diccionario que almacenara la informacion de la cantidad vendida en cada Estado a lo largo de los meses
ventasEstado = {}

for mes in MESES:

  ventasMes = dataframes[mes].groupby('Estado')['Cantidad Pedida'].sum()
  estados = ventasMes.reset_index(inplace=False)['Estado'].values

  for estado in estados:
    if not(estado in ventasEstado):
      ventasEstado[estado] = {'Enero':0, 'Febrero':0, 'Marzo':0, 'Abril':0, 'Mayo':0, 'Junio':0, 'Julio':0, 'Agosto':0, 'Septiembre':0, 'Octubre':0, 'Noviembre':0, 'Diciembre':0}
      ventasEstado[estado][mes] = ventasMes.loc[estado]
    else:
      ventasEstado[estado][mes] = ventasMes.loc[estado]

fig = go.Figure()

# Creamos un grafico por cada Estado
for estado in ventasEstado:
  fig.add_trace(go.Scatter(
    x=MESES,
    y=list(ventasEstado[estado].values()),
    mode='lines',
    name=estado,
    line=dict(width=2, shape='spline'),
  ))

fig.update_layout(
  title='Unidades vendidas de cada Estado por mes',
  xaxis=dict(title='Meses', tickvals=MESES),
  yaxis=dict(title='Unidades vendidas', side='left'),
  legend=dict(y=0.3, yanchor='bottom'),
)

fig.show()
```

## 4. Análisis del producto más vendido:
* ¿Cuál es el producto más vendido en general y en cada mes?
* ¿Qué factores crees que han contribuido al éxito de ese producto en particular?

```python
# Creamos dos listas, una que almacenara los nombres de los productos mas vendidos por mes y otra su cantidad
masVendidoMesesNombre = []
masVendidoMesesCantidad = []

for mes in MESES:
  masVendidoMes = dataframes[mes].groupby('Producto')['Cantidad Pedida'].sum()
  masVendidoMes = masVendidoMes.reset_index(inplace=False).sort_values(by='Cantidad Pedida', ascending=False).reset_index(drop=True)
  # masVendidoMesesNombre[mes] = {'Producto': masVendidoMes.iloc[0]['Producto'], 'Cantidad': masVendidoMes.iloc[0]['Cantidad Pedida']}
  masVendidoMesesNombre.append(masVendidoMes.iloc[0]['Producto'])
  masVendidoMesesCantidad.append(masVendidoMes.iloc[0]['Cantidad Pedida'])

# Creamos un diccionario donde almacenaremos el total de cada producto mas vendido de cada mes
mayoresVentas = {}

for i in range(len(masVendidoMesesNombre)):
  if not(masVendidoMesesNombre[i] in mayoresVentas):
    mayoresVentas[masVendidoMesesNombre[i]] = masVendidoMesesCantidad[i]
  else:
    mayoresVentas[masVendidoMesesNombre[i]] += masVendidoMesesCantidad[i]

# Creamos una lista donde el primer valor va a ser la cantidad vendida del producto mas vendido en general y el segundo su nombre
masVendido = [0,'']

for producto in mayoresVentas:
  if mayoresVentas[producto] > masVendido[0]:
    masVendido[0] = mayoresVentas[producto]
    masVendido[1] = producto

fig = go.Figure()

fig.add_trace(go.Bar(
  x = MESES,
  y= masVendidoMesesCantidad,
  name = 'Unidades Vendidas',
  marker = dict(color= 'Light Blue'),
  text=masVendidoMesesNombre,
))

# Añadimos una anotacion para especificar el producto mas vendido de todos en el grafico
fig.add_annotation(
  text=f'Producto mas vendido en general: {masVendido[1]} con {masVendido[0]} unidades en total',
  xref='paper',
  yref='paper',
  x=0,
  y=1,
  showarrow=False
)

fig.update_layout(
  title = 'Producto mas vendido por Mes',
  xaxis = dict(title='Productos'),
  yaxis = dict(title='Unidades')
)

fig.show()
```

```python
dfConcatGroupbyDiaMes = dataframesConcat.groupby(by=['Mes', 'Dia'])['Cantidad Pedida'].sum()
dfConcatGroupbyDiaMes = dfConcatGroupbyDiaMes.reset_index(inplace=False).sort_values(by=['Mes', 'Dia'], ascending=[True,True]).reset_index(drop=True)

dfConcatGroupbyDiaMes['Promedio Ventas Mensual'] = dfConcatGroupbyDiaMes['Mes'].map(lambda x: promedios_ventas_mensuales[x-1])

dfConcatGroupbyDiaMes['FechaUnida'] = '2019'+'-'+ dfConcatGroupbyDiaMes['Mes'].astype(str)+'-'+dfConcatGroupbyDiaMes['Dia'].astype(str)
dfConcatGroupbyDiaMes['FechaUnida'] = pd.to_datetime(dfConcatGroupbyDiaMes['FechaUnida'])

dfConcatGroupbyDiaMes
```

```python
# Definir días previos
dias_previos = 3

# Definir fechas festivas
fechasFestivas = [datetime.strptime('01-01-2019', '%d-%m-%Y'), datetime.strptime('21-01-2019', '%d-%m-%Y'),
                      datetime.strptime('22-02-2019', '%d-%m-%Y'),datetime.strptime('27-05-2019', '%d-%m-%Y'),
                      datetime.strptime('19-06-2019', '%d-%m-%Y'),datetime.strptime('04-07-2019', '%d-%m-%Y'),
                      datetime.strptime('02-09-2019', '%d-%m-%Y'),datetime.strptime('12-10-2019', '%d-%m-%Y'),
                      datetime.strptime('11-11-2019', '%d-%m-%Y'),datetime.strptime('18-11-2019', '%d-%m-%Y'),
                      datetime.strptime('25-12-2019', '%d-%m-%Y')]

# Listas para almacenar resultados
ventasTotalesPeriodoMes = []

fecha_festiva_previa = []
for fecha in fechasFestivas:
  ventas_periodo = 0
  promedio_ventas = 0
  fecha_anterior = fecha - timedelta(days=dias_previos)  # Calcular fecha previa
  fecha_posterior = fecha + timedelta(days=1)  # Calcular fecha previa
  fecha_festiva_previa.append(fecha_anterior)
  # Filtrar ventas dentro del período
  ventas_periodo = dataframesConcat[(dataframesConcat['Fecha de Pedido'] >= fecha_anterior) & (dataframesConcat['Fecha de Pedido'] < fecha_posterior)]
  ventas_periodo = ventas_periodo['Cantidad Pedida'].sum()
  # Calcular promedio y agregar a la lista
  promedio_ventas = ventas_periodo / dias_previos
  ventasTotalesPeriodoMes.append(promedio_ventas)
```

```python
festivos_data = {
    'nombresFestivos': nombresFestivos,  # Lista de nombres de festivos
    'fecha_previa': fecha_festiva_previa,
    'fecha_festiva': fechasFestivas,  # Fechas festivas
    'promedio periodo festivo':ventasTotalesPeriodoMes
}

# Crea un DataFrame para almacenar los resultados.
resultados = pd.DataFrame(columns=['Fecha Festiva', 'Total Cantidad Pedida'])
# Crear el DataFrame
df_combinado = pd.DataFrame(festivos_data)
df_combinado
```
